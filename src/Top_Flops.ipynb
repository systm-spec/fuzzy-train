```python
import pandas as pd
import matplotlib.pyplot as plt

pokemon_dataset = pd.read_csv('../src/data/pokedex_420.csv')

pkmn_global = pokemon_dataset.drop(['Unnamed: 0','name', 'japanese_name', 'catch_rate', 'base_experience',
                                    'base_friendship', 'growth_rate', 'egg_cycles', 'egg_type_1', 'egg_type_2',
                                    'egg_type_number', 'percentage_male'],axis=1)

pkmn = pkmn_global[pkmn_global['generation'] <= 3]
```

```python
pkmn
```

```python
groeste_pkmn = pkmn[['german_name', 'height_m']]
groesten_fuenf = groeste_pkmn.sort_values(by=['height_m'], ascending=False)
top_5_groeste = groesten_fuenf.head(5)
top_5_groeste
```

```python
plt.figure(figsize=(6,4))
plt.bar(top_5_groeste["german_name"], top_5_groeste["height_m"], color='orange', ec='black')
plt.title("Die fünf größten Pokémon", fontsize=14, fontweight="bold")
plt.xlabel("Pokémon (Deutscher Name)", fontsize=12)
plt.ylabel("Grösse (m)", fontsize=12)
plt.xticks(fontsize=10, rotation=45)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
